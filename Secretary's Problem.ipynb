```python
## Importing packages
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
n = 100
# candidates = np.random.randint(low = 0, high = n-1, size = n)   ## with repeat 
candidates = np.random.choice(range(1, n+1), n, replace=False)   ## without repeat
itr = 10000

probs = []
for k in range(5, 101, 5):
  best_chosen_candidate = []
  for i in range(itr):
    best_chosen_candidate.append(Choose_Best(candidates, k))
    np.random.shuffle(candidates)
  
  best_chosen_candidate = np.array(best_chosen_candidate)
  prob = sum(best_chosen_candidate == max(best_chosen_candidate))/itr
  probs.append(prob)
```

```python
## PLotting the graph:
plt.scatter(range(5, 101, 5), probs)
plt.xlabel('Number of rejected candidates (k)')
plt.ylabel('Probability of choosing the best candidate')
plt.show()
```

```python
def Choose_Best(candidates, k):
  best_rejected = np.max(candidates[:k])
  remained = candidates[k:]
  if (sum(remained > best_rejected) != 0):
    best_candidate = remained[remained > best_rejected][0]
  else:
    best_candidate = candidates[-1]
  return best_candidate
```

```python
probs = []
for n in range(3, 101):
  candidates = np.random.choice(range(1, n+1), n, replace=False) 
  k = n/np.exp(1)
  best_chosen_candidate = []
  for i in range(itr):
    best_chosen_candidate.append(Choose_Best(candidates, int(round(k))))
    np.random.shuffle(candidates)

  best_chosen_candidate = np.array(best_chosen_candidate)
  prob = sum(best_chosen_candidate == max(best_chosen_candidate))/itr
  probs.append(prob)
```

```python
## PLotting the graph:
n_list = np.arange(3,101)
plt.scatter(n_list, probs)
plt.xlabel('n')
plt.ylabel('Probability of choosing the best candidate')
plt.show()
```

```python
import pandas as pd    
import seaborn as sns 
data = pd.read_csv("wine.csv")  
print(data.head())   
```

**Plotting the distribution of the 12th column:**

```python
sns.distplot(data['12'])
```

**Plotting the means for different populations and reporting the means and standard deviations:**

```python
import warnings
warnings.filterwarnings("ignore")
population = data['12']
Mean = np.mean(population)
std = np.std(population)
n_list = [10, 20, 30, 40, 50, 70, 90, 100, 10, 130, 150, 170]
s = 1000
plt.figure(figsize=(25,15))
for i in range(len(n_list)):
  avg = []
  for j in range(s):
    selected_population = np.random.choice(data['12'], n_list[i], replace=False)
    avg.append(np.mean(selected_population))
  Mean_CLT = np.mean(avg)
  std_CLT = np.std(avg)
  print('Mean = ', Mean)
  print('Std = ', std/np.sqrt(n_list[i]))
  print('Mean from CLT = ', Mean_CLT)
  print('Std from CLT = ', std_CLT)
  print('................................................')
  plt.subplot(3,4,i+1)
  sns.distplot(avg)
  plt.title('n = {}, Mean = {}, std = {}'.format(n_list[i], round(Mean_CLT,4), round(std_CLT,4)))
```
